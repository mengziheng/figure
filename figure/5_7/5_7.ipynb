```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams

rcParams["font.size"] = 24
rcParams["font.family"] = "Times New Roman"
rcParams["text.usetex"] = True
rcParams["font.family"] = "serif"
rcParams["font.serif"] = ["Times New Roman"] + rcParams["font.serif"]
# 数据
a = [0.68,0.73,1.05,1.02,0.78,0.71,0.66]
b = [2.10,2.14,2.71,2.21,2.59,1.51,1.29]
c = [24,25,26,27,28,29,30]
x = np.array(c)  # x轴的位置直接使用c

# 创建图形和一个子图
fig, ax1 = plt.subplots(figsize=(6, 4))

color_blue = "#72b3d0"  # BCCE 的颜色
color_green = "#72d072"  # baseline 的颜色

# 绘制a的柱状图
ax1.bar(x - 0.2, a, width=0.4, label='Average', color=color_blue, hatch="//", edgecolor="black", alpha=0.7)
ax1.set_ylabel('Metric(%)', color='black')
ax1.tick_params(axis='y', labelcolor='black')

# 绘制b的柱状图
ax1.bar(x + 0.2, b, width=0.4, label='Max', align='center', color = color_green, hatch='\\\\', edgecolor="black",alpha=0.7)

# 设置x轴的刻度标签为c的值
ax1.set_xticks(c)
ax1.set_xticklabels(c)
ax1.set_xlabel('log$_2$(n)')

ax1.set_ylim(0, 5)
ax1.set_ylabel('Metric(\%)', color='black')
ax1.tick_params(axis='y', labelcolor='black')

plt.legend(loc="upper right")

plt.savefig("5_7.pdf", bbox_inches = "tight")
plt.show()
plt.close()
```
