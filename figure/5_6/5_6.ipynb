```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams["font.size"] = 24
rcParams["text.usetex"] = True
rcParams["font.family"] = "serif"
rcParams["font.serif"] = ["Times New Roman"] + rcParams["font.serif"]

file_path = "../../output/5_8/5_8.xlsx"
df = pd.read_excel(file_path)

# 提取数据
gpu_nums = np.array([1, 2, 4, 8])  # 确保横坐标是等比例的
in_memory_times = df.iloc[0, 1:].values  # 第一行是 in-memory 版本
out_memory_times = df.iloc[1, 1:].values  # 第二行是 out-of-memory 版本

# 计算柱子宽度和间距
bar_width = 0.3  # 设置柱子宽度
x_positions = np.arange(len(gpu_nums))  # 生成等距的横坐标

color_blue = "#72b3d0"  # BCCE 的颜色
color_green = "#72d072"  # baseline 的颜色

plt.figure(figsize=(6, 4))
plt.bar(x_positions - bar_width/2, in_memory_times, width=bar_width, label="In Memory", color=color_blue, hatch="//", edgecolor="black", alpha=0.7)
plt.bar(x_positions + bar_width/2, out_memory_times, width=bar_width, label="Out Memory", color = color_green, hatch='\\\\', edgecolor="black",alpha=0.7)

# 设置横轴刻度，使其等间距
plt.xticks(x_positions, gpu_nums)  

# 设置坐标轴标签
plt.xlabel("Number of GPUs")
plt.ylabel("Latency (unit:ms)")

# 添加图例
plt.legend(fontsize=16)

plt.savefig("5_6.pdf", bbox_inches = "tight")
plt.show()
plt.close()

```
